```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

excel_file_ga = 'sample_data/data - ga.xlsx'
excel_file_gan = 'sample_data/data - gan.xlsx'
excel_file_mc = 'sample_data/data - mc.xlsx'
excel_file_mc_original = 'sample_data/data - smb.xlsx'

df = pd.read_excel(excel_file_ga)
dff = pd.read_excel(excel_file_gan)
dfff = pd.read_excel(excel_file_mc)
dffff = pd.read_excel(excel_file_mc_original)


fifth_column_ga = df.iloc[:, 6]
fifth_column_gan = dff.iloc[:, 6]
fifth_column_mc = dfff.iloc[:, 6]
fifth_column_mc_original = dffff.iloc[:, 6]

data = pd.DataFrame({
    'GA': fifth_column_ga,
    'GAN': fifth_column_gan,
    'MC': fifth_column_mc,
    'SMB': fifth_column_mc_original
})


palette = {
    'GA': 'green',
    'GAN': 'purple',
    'MC': 'blue',
    'SMB': 'orange'
}


plt.figure(figsize=(8, 8))


ax = sns.boxplot(data=data, palette=palette, width=0.3)

plt.ylabel('Lenience Values')

ax.grid(True)
plt.show()
```

Corner plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# Leer el archivo de datos
excel_file = 'sample_data/data - mc.xlsx'  # Cambia esto al nombre de tu archivo de Excel
df = pd.read_excel(excel_file)

# Seleccionar todas las columnas excepto la primera
df_selected = df.iloc[:, 1:]

# Crear el pairplot
pairplot = sns.pairplot(df_selected, plot_kws={'alpha': 0.5}, corner=True)

# Función para agregar la correlación de Spearman y el valor p en los gráficos
def add_spearman_correlation(x, y, **kwargs):
    corr, p_value = spearmanr(x, y)
    ax = plt.gca()
    ax.annotate(f'spearman = {corr:.2f}\np = {p_value:.2e}',
                xy=(0.1, 0.9), xycoords=ax.transAxes, fontsize=10, color='red')

# Aplicar la función a cada gráfico en el pairplot
pairplot.map_lower(add_spearman_correlation)

# Mostrar el gráfico
plt.show()
```
