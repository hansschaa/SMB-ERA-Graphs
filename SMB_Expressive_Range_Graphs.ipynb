```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

col1 = 'Empty Spaces'
col2 = 'Significant Jumps'

excel_file_ga = 'sample_data/data - ga.xlsx'
excel_file_gan = 'sample_data/data - gan.xlsx'
excel_file_mc = 'sample_data/data - mc.xlsx'
excel_file_mc_original = 'sample_data/data - mc_original.xlsx'

# Leer los archivos de datos
df = pd.read_excel(excel_file_ga)
dff = pd.read_excel(excel_file_gan)
dfff = pd.read_excel(excel_file_mc)
dffff = pd.read_excel(excel_file_mc_original)

df_selected = df[[col1, col2]]
dff_selected = dff[[col1, col2]]
dfff_selected = dfff[[col1, col2]]
dffff_selected = dffff[[col1, col2]]

# Crear el gráfico de hexágonos
fig, ax = plt.subplots(figsize=(8, 8))

# Hexbin plot para cada conjunto de datos con diferentes gridsize (ancho y alto)
hb = ax.hexbin(df_selected[col2], df_selected[col1], gridsize=(21, 3), mincnt=1, cmap='Greens', vmin=0, vmax=15)
hbb = ax.hexbin(dff_selected[col2], dff_selected[col1], gridsize=(26, 2), mincnt=1, cmap='Purples', vmin=0, vmax=10)
hbbb = ax.hexbin(dfff_selected[col2], dfff_selected[col1], gridsize=(30, 15), mincnt=2, cmap='Blues', vmin=0, vmax=10)
hbbbb = ax.hexbin(dffff_selected[col2], dffff_selected[col1], gridsize=(45,8), mincnt=1, cmap='Oranges', vmin=0, vmax=2)

# Establecer límites en los ejes
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

# Establecer la misma proporción en ambos ejes
ax.set_aspect('equal')

# Añadir cuadrícula y ajustar marcas en los ejes
ax.grid(True)
ax.set_xticks([i/10 for i in range(11)])  # Marcas en el eje x cada 0.1
ax.set_yticks([i/10 for i in range(11)])  # Marcas en el eje y cada 0.1
# Añadir etiquetas y leyenda
ax.set_xlabel(col2, fontsize=10)
ax.set_ylabel(col1, fontsize=10)
ax.set_title('Hexbin Plot: ' + col1 + ' vs ' + col2)

# Añadir una leyenda con un icono de color
legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=10, label='GA'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='purple', markersize=10, label='GAN'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='MC'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='orange', markersize=10, label='SMB')
]
ax.legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1, 1), borderaxespad=0.)

plt.tight_layout()
plt.show()
```
